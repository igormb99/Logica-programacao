```python
#Simulação de automatização de mailing
#For

#lista de mailing de clientes
lista_de_clientes = ["cliente1@email", "cliente2@email", "cliente3@email"]

#Envio de e-mail para cada cliente
for email in lista_de_clientes:
  print("Enviando e-mail para: " + email)
  print("E-mail enviado com sucesso!")
```

```python
#Simulação de automação de estoque
estoque = 10

#Simulando vendas de estoque
while estoque > 0:
  print(f"estoque final: {estoque}")
  print ("realizando venda...")
  estoque -= 1 #reduzindo estoque
  print("venda concluída!")

print("estoque esgotado! pausando vendas")
```
